<h1>Exemplo personalizado de superfície</h1>

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parâmetros 
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, 2 * np.pi, 200)
u, v = np.meshgrid(u, v)
```

```python
# paramétricas
x = np.cos(u) * np.cos(v)
y = np.cos(u) * np.sin(v)
z = np.sin(u)
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)

# ticks
ax.set_xlabel('Eixo x')
ax.set_ylabel('Eixo y')
ax.set_zlabel('Eixo z')

# mostra gráfico
plt.show()
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)

# ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([-1,-0.5,0,0.5,1])

# mostra gráfico
plt.show()
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)

# intervalos em cada eixo
ax.set_xlim3d(-2, 2)
ax.set_ylim3d(-2, 2)
ax.set_zlim3d(-1, 1)

# mostra gráfico
plt.show()
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)

# título
ax.set_title("ESFERA")

# mostra gráfico
plt.show()
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)

# ângulo de apresentação
ax.view_init(30,90)

# mostra gráfico
plt.show()
```

```python
# gráfico
ax= plt.axes(projection='3d')

ax.plot_surface(
    x, y, z,
    alpha=0.3, # tranparência
    cmap='gist_heat' # cor             
)

# mostra gráfico
plt.show()
```
