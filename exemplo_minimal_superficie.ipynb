<h1>Exemplo mínimo de superfície</h1>

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parâmetros 
u = np.linspace(0, 2 * np.pi, 200)
v = np.linspace(0, 2 * np.pi, 200)
u, v = np.meshgrid(u, v)
```

```python
# paramétricas
x = np.cos(u) * np.cos(v)
y = np.cos(u) * np.sin(v)
z = np.sin(u)
```

```python
# gráfico
ax= plt.axes(projection='3d')
ax.plot_surface(x, y, z)
plt.show()
```
