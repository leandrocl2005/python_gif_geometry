<h1>Animando tube Plot</h1>

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
num_frames = 150
for step in range(0, num_frames):
    
    # janela de figura
    plt.figure(figsize=(10,10))
    
    # parmetros 
    u= np.linspace(0, 8 * np.pi/num_frames * step ,30)
    v= np.linspace(0, 2 * np.pi, 30)
    
    # gride de parmetros
    u, v= np.meshgrid(u, v)
    
    # paramtricas
    x = np.cos(v)
    y = np.sin(v) + np.cos(u) - 1
    z = u
    ax= plt.axes(projection='3d')
    
    # grfico
    ax.plot_surface(x, y, z, alpha=0.6, cmap='gist_heat')
    
    # legendas
    ax.set_xlabel ( 'Eixo x')
    ax.set_ylabel( 'Eixo y')
    ax.set_zlabel ( 'Eixo z')
    
    # intervalos em cada eixo
    ax.set_xlim3d(-2, 2)
    ax.set_ylim3d(-4, 4)
    ax.set_zlim3d(-0.2, 8 * np.pi)
    
    # ttulo
    plt.title('Tube plot')
    
    # limpa o display
    clear_output(wait=True)
    
    # ngulo
    ax.view_init(30,30)
    
    # nome do arquivo
    filename = "testesFig/fig" + str(step) + ".png"
    
    # salva imagem
    plt.savefig(filename, dpi=96)
    
    # mostra no display
    plt.show()
```
