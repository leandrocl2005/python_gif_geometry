<h1>Animanado esferas</h1>

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gráfico
u = np.linspace(0, 2 * np.pi, 30)
v = np.linspace(0, 2 * np.pi, 30)
u, v = np.meshgrid(u, v)
x = np.cos(u) * np.cos(v)
y = np.cos(u) * np.sin(v)
z = np.sin(u)
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z)
    
# nome do arquivo
nome_arquivo = "esfera.png"
    
# salva imagem
plt.savefig(nome_arquivo, dpi=96)
```

```python
total = 3
for num in range(0, total):
    
    # Gráfico
    u = np.linspace(0, 2 * np.pi, 30)
    v = np.linspace(0, 2 * np.pi, 30)
    u, v = np.meshgrid(u, v)
    x = np.cos(u) * np.cos(v)
    y = np.cos(u) * np.sin(v)
    z = np.sin(u)
    ax = plt.axes(projection='3d')
    ax.plot_surface(x, y, z)

    # nome do arquivo
    nome_arquivo = "esfera" + str(num) + ".png"

    # salva imagem
    plt.savefig(nome_arquivo, dpi=96)
```

```python
total = 100
for num in range(0, total):
    # Gráfico
    u = np.linspace(0, 2 * np.pi / total * num, 30)
    v = np.linspace(0, 2 * np.pi, 30)
    u, v = np.meshgrid(u, v)
    x = np.cos(u) * np.cos(v)
    y = np.cos(u) * np.sin(v)
    z = np.sin(u)
    ax = plt.axes(projection='3d')
    ax.plot_surface(x, y, z)
    ax.set_xlim3d(-2, 2)
    ax.set_ylim3d(-2, 2)
    ax.set_zlim3d(-2, 2)

    # nome do arquivo
    nome_arquivo = "esfera" + str(num) + ".png"

    # salva imagem
    plt.savefig(nome_arquivo, dpi=96)
```
